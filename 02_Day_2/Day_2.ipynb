```python
import numpy as np
```

```python
report_list = open("report_list.txt", "r").read()
report_list = report_list.split("\n")[:1000]

for i, report in enumerate(report_list) :
    report_list[i] = list(map(int,report.split(" ")))
```

```python
### Verifiy safety conditions - PART 1

# Function that checks monotonicity of a report
def monotonicity_check(report):
    increasing_report = np.sort(report).tolist()
    decreasing_report = increasing_report.copy()
    decreasing_report.reverse()
    monotonicity = bool(np.logical_or(increasing_report == report, decreasing_report == report)) # monotonicity check
    return monotonicity

# Function that checks that the slope of a report is correct
def slope_check(report):
    for i in range(len(report)-1):
        slope = abs(report[i+1] - report[i])
        if slope == 0 or slope > 3:
            return False
    return True
```

```python
#Initialize
report_safety_list = [] # List of boolean, True if the report is safe, False otherwise

## Verify monotonicity of the reports
report_monotonic_check_list = [] # List of boolean, True if the report is monotonic, False otherwise
for report in report_list:
    monotonicity = monotonicity_check(report)
    report_monotonic_check_list.append(monotonicity)

report_safety_list = report_monotonic_check_list

## Verify slope of the monotonic reports 
slope_check_index = np.where(report_safety_list)[0].tolist()

for i in slope_check_index : 
    report_safety_list[i] = slope_check(report_list[i])

## Compute number of safe reports
n_safe_report = np.sum(report_safety_list)
print("Number of safe reports : ", n_safe_report)
```

```python
### Verifiy safety conditions with Problem Dampener - PART 2

# Locate unsafe reports
report_safety_list = np.array(report_safety_list)
unsafe_report_index = np.where(report_safety_list == False)[0].tolist()

# Check if the reports become safe when getting rid of one of the values
for i in unsafe_report_index:
    for j in range(len(report_list[i])): 
        report = report_list[i].copy()
        report.pop(j)
        # Check safety of the report where one value has been poped
        safety = monotonicity_check(report) and slope_check(report)
        # Update the safety of the report if it became safe
        if safety :
            report_safety_list[i] = True
            break

## Compute number of safe reports
n_safe_report = np.sum(report_safety_list)
print("Number of safe reports with the Problem Dampener: ", n_safe_report)
    
```
