```python
import numpy as np
```

```python
### Import the word_search and convert it into an array
raw_word_search = open("word_search.txt", "r").read().split('\n')[:-1]
for i, word_search_line in enumerate(raw_word_search) :
    raw_word_search[i] = list(word_search_line)
word_search = np.array(raw_word_search)

print(word_search)
```

```python
### Part 1 - Count the number of 'XMAS' in the word search

# Function that count the number of 'word' appearances in an 1D-array of characters
def search_word(line, word):

    n_words = 0 # Initialize the count
    reversed_word = word[::-1]
    if len(line) >= len(word) : # Make sure that the string is long enough to contain the word

        # Search the word or the reversed word over the whole string
        for i in range(len(line)- len(word) + 1):    # The last index at which the word can begin is -4
            if ''.join(line[i:i+len(word)]) == word or ''.join(line[i:i+len(word)]) == reversed_word:
                n_words += 1

    return n_words
```

```python
# Function that extract the diagonals and the antidiagonals of a 2D-square-array as a list of 1D-arrays
def extract_diagonals(word_search):
    diagonals = []
    flipped_lr_word_search = np.fliplr(word_search)
    _ , n_columns = np.shape(word_search)
    for i in range(n_columns) : 
        diagonals.append(np.diagonal(word_search,i).copy())
        diagonals.append(np.diagonal(flipped_lr_word_search,i).copy())
        if i != 0 :
            diagonals.append(np.diagonal(word_search,-i).copy())
            diagonals.append(np.diagonal(flipped_lr_word_search,-i).copy())
    return diagonals
```

```python
## Count the number of 'XMAS' in the word search

# Initialize the count
n_words = 0

# Extract the diagonals of the word search
diagonals = extract_diagonals(word_search)

# Horizontal search
for row in word_search:
    n_words += search_word(row, word='XMAS')

# Vertical search
for column in word_search.T:
    n_words += search_word(column, word='XMAS')

# Diagonal search
for diagonal in diagonals :
    n_words += search_word(diagonal, word='XMAS')

print("Number of XMAS in the word search :", n_words)
```

```python
### Part 2 - Count number of X-"MAS" in the word search

# Function that find the list of central index of a odd length 'word' in an 1D-array of characters
def search_word_indices(line, word):
    indices = []

    if len(line) >= len(word) : # Make sure that the string is long enough to contain the word
        reversed_word = word[::-1]

        # Search the word or the reversed word over the whole string
        for i in range(len(line) - len(word) + 1):    # The last index at which the word can begin is -len(word)
            if ''.join(line[i:i+len(word)]) == word or ''.join(line[i:i+len(word)]) == reversed_word:
                indices.append(i+len(word)//2)
    
    return indices
```

```python
word = 'MAS'
n_words = 0

coordinates_list = []
n_rows , n_columns = np.shape(word_search)

# Search the word through mid and upper diagonals
for offset in range(n_columns):
    diagonal = np.diagonal(word_search, offset)     # Extract the offset diagonal
    indices_list = search_word_indices(diagonal, word)  # Extract the indices of the words' middle letter in the diagonal
    for index in indices_list :
        coordinates_list.append([index,index+offset])   # Save the coordinates of the word's middle letter
        
# Search the word through lower diagonals
for offset in range(1,n_rows) :
    diagonal = np.diagonal(word_search, -offset)    # Extract the -offset diagonal
    indices_list = search_word_indices(diagonal, word)  # Extract the indices of the words' middle letter in the diagonal
    for index in indices_list:
        coordinates_list.append([index+offset,index])   # Save the coordinates of the word's middle letter

# Iterate over each coordinate
for coordinate in coordinates_list:
    x, y = coordinate
    anti_word = word_search[x+1, y-1] + word_search[x, y] + word_search[x-1, y+1]   # Construct the word in the antidiagonal centered at the coordinate x, y
    if anti_word == word or anti_word == word[::-1] :   # Check if the antidiagonal word is correct
        n_words += 1

print("Number of X-'MAS' in the word search :", n_words)
```
